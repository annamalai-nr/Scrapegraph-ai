```python
# Define the configuration for the FetchNode
fetch_node_config = {
    "headless": False,  # Set to False to observe browser actions; True for headless mode
    "verbose": True,
    "use_playwright": True,  # Enable Playwright integration
    "playwright_options": {
        "java_script_enabled": True,  # Ensure JavaScript is enabled
        "cookies": [
            # Add any specific cookies if required
        ],
        # Additional Playwright configurations can be added here
    }
}

# Create the FetchNode instance
fetch_node = FetchNode(
    input="url",
    output=["fetched_content"],
    node_config=fetch_node_config
)

# Define the configuration for the ParseNode
parse_node_config = {
    "verbose": True,
    "parse_html": True,  # Enable HTML parsing
    "chunk_size": 4096,  # Define the chunk size for parsing
}

# Create the ParseNode instance
parse_node = ParseNode(
    input="fetched_content",
    output=["parsed_content"],
    node_config=parse_node_config
)
```

```python
def no_llm_scrape(url):
    # Define the initial state with the target URL
    state = {
        "url": url  # Replace with your target URL
    }

    # Execute the FetchNode to fetch the content
    state = fetch_node.execute(state)

    # Execute the ParseNode to parse the fetched content
    state = parse_node.execute(state)

    # Retrieve the parsed content
    parsed_content = state.get("parsed_content")
    return parsed_content[0]
```

```python
scraped_content = []
time_taken = []
for url in tqdm(urls):
    start_time = time()
    content = no_llm_scrape(url)
    end_time = time()
    time_taken.append(end_time - start_time)
    scraped_content.append(content)
```
